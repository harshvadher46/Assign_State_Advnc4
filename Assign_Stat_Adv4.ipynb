```python
'''
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
 
Ans:
Both t-tests and z-tests are statistical methods used to make inferences about population parameters based on sample data, 
but they are applied under different circumstances due to differences in assumptions about the population and sample  

T-test:
--> A t-test is used when the sample size is small (typically less than 30) or when the population standard deviation is unknown.
--> It is particularly useful when dealing with small sample sizes because it incorporates an adjustment for the greater uncertainty that comes with estimating population parameters from limited data.
--> 

Z-test:
--> A z-test is used when the sample size is large (typically greater than 30) and when the population standard deviation is known.
--> It assumes that the sample comes from a normally distributed population.

'''
```

```python
'''
Q2: Differentiate between one-tailed and two-tailed tests.
--> One-tailed and two-tailed tests are types of hypothesis tests that differ in their directionality of the hypothesis
and the way they evaluate significance:

One-tailed test:

--> In a one-tailed test, the null hypothesis specifies a particular direction (either greater than or less than), and the alternative hypothesis focuses on that direction exclusively.
--> The critical region, where the observed test statistic leads to rejection of the null hypothesis, is located entirely in one tail of the distribution.
--> One-tailed tests are used when there is a specific directional hypothesis or when researchers are only interested in deviations from the null hypothesis in one direction.
--> Example: Testing whether a new drug increases average reaction times. The one-tailed null hypothesis would be that the drug has no effect or decreases reaction times,
    and the alternative hypothesis would be that the drug increases reaction times.

Two-tailed test:
--> In a two-tailed test, the null hypothesis does not specify a direction,
    it merely states that there is no significant difference or relationship.
--> The alternative hypothesis is typically non-directional, suggesting that there is a significant difference or relationship,
    but without specifying the direction.
--> Two-tailed tests are used when researchers are interested in deviations from the null hypothesis in either direction
    or when the research question is not specific about the direction of the effect.
--> Example: Testing whether a coin is fair. The null hypothesis would be that the coin is fair (i.e., has a 50% chance of landing heads), 
             and the alternative hypothesis would be that the coin is biased (i.e., not 50% chance of landing heads), 
             without specifying whether it is biased towards heads or tails.
'''
```

```python
'''
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Type I error:

--> Also known as a "false positive," a Type I error occurs when the null hypothesis is incorrectly rejected when it is actually true. 
    In other words, it is the error of concluding that there is a significant effect or difference when there isn't one.
--> The probability of committing a Type I error is denoted by the significance level, often denoted as α (alpha). 
    It represents the probability of rejecting the null hypothesis when it is true.
--> Type I errors are typically more concerning in situations where making incorrect conclusions can have serious consequences or costs.
--> Example scenario: Suppose a medical test for a rare disease has a false-positive rate of 5%. 
    If a healthy person is incorrectly diagnosed with the disease based on the test result, it would be a Type I error.

Type II error:

--> Also known as a "false negative," a Type II error occurs when the null hypothesis is incorrectly not rejected when it is actually false. In other words,
    it is the error of failing to detect a significant effect or difference when there is one.
--> The probability of committing a Type II error is denoted by β (beta). It represents the probability of accepting the null hypothesis when it is false.
--> Type II errors are often associated with situations where the sample size is too small, the effect size is too small, or the variability in the data is too high.
--> Example scenario: Imagine a new drug is developed to lower blood pressure. If the drug is effective at reducing blood pressure, but a clinical trial fails to detect this effect due to
                      limitations in the study design or insufficient statistical power, it would be a Type II error.

'''
```

```python
'''
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

--> A confidence interval is a range of values that is used to estimate the true value of a population parameter, such as the mean, with a certain level of confidence. It provides a measure of the uncertainty associated with the estimate. 
The confidence level represents the probability that the interval contains the true population parameter.
Here's how to calculate a confidence interval for a population mean,μ, using the sample mean,and the sample standard deviation.


1) - Choose a confidence level. Common choices include 90%, 95%, or 99% confidence.
2) - Determine the appropriate critical value (z-score or t-score) for the chosen confidence level. This value depends on the sample size and the desired confidence level. For large sample sizes (typically > 30) 
     or when the population standard deviation is known, you use the z-score from the standard normal distribution. 
     For smaller sample sizes (typically ≤ 30) or when the population standard deviation is unknown, you use the t-score from the t-distribution.
3) - Calculate the margin of error, which is the product of the critical value and the standard error of the mean:
  Margin of error = Critical Value * s/underoot n
  
  where,
  s is the sample standard deviation,
  n is the sample size.
Construct the confidence interval by adding and subtracting the margin of error from the sample mean:

Confidence Interval= 
Sample means ± Margin of Error

EXAMPLE USAGE :::
Suppose you want to estimate the average height of students in a school. You take a random sample of 50 students and measure their heights.
The sample mean height is 65 inches, and the sample standard deviation is 3 inches.

1:-Choose a 95% confidence level.

2:-Determine the critical value. Since the sample size is large (n > 30), we'll use the z-score for a 95% confidence level,
   which is approximately 1.96.
   
3:-Calculate the margin of error: Margin Of Error = 0.83 

4:-Construct the confidence interval:

Confidence Interval=65 ± 0.83 =(64.17,65.83)

So, we can be 95% confident that the true average height of students in the school falls within the range of 64.17 inches to 65.83 inches.
'''
```

```python
'''
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

NUll hypothesis (H0) :- the weight lost drug is not significantly effective :the mean weight loss is not different from zero.
Alternate Hypothesis(H1) :- The weight loss drug is significantly effective :the mean weight loss is different from zero.

We will use a two-tailed t-test since we want to determine if the mean weight loss is significantly different from zero.
Given:
sample size = 50 
sample mean = 6 pound
sample std = 2.5pound
CI = 95%

We need to calculate the t-statistic and compare it with the critical t-value at the 95% confidence level with 
n-1 degree of freedom (df=49)

t = 16.99(By using formula)
Now, we need to find the critical t-value for a two-tailed test at the 95% confidence level with 49 degrees of freedom.

Using a t-table or statistical software, the critical t-value is approximately ±2.01.

Since our calculated t-statistic (16.99) is much larger than the critical t-value (2.01), we reject the null hypothesis.

Conclusion:
At the 95% confidence level, there is sufficient evidence to conclude that the weight loss drug is significantly effective, 
as the mean weight loss is significantly different from zero
'''
```

```python
# Q-11
                    # Answer

import numpy as np

# Sample size
n = 500

# Sample proportion satisfied
p_hat = 0.65

# Confidence level
confidence_level = 0.95

# Significance level
alpha = 1 - confidence_level

# Calculate z-score
z_score = np.quantile(np.random.standard_normal(size=10000), 1 - alpha / 2)

# Calculate margin of error
margin_of_error = z_score * np.sqrt(p_hat * (1 - p_hat) / n)

# Calculate confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print(f"95% confidence interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```
